```python
df.columns=["Table name","State code","Distt code","Area name","Total/Rural/Urban",
           "Education level","total person","total males","total females","total main worker",
           "males main worker","females main worker","total marginal worker","males marginal worker",
            "females marginal worker","total MAW","males MAW","females MAW","total non worker",
            "males non worker","females non worker","total unemployed","males unemployed","females unemployed"]
```

```python
df.head()
```

```python
df=df.drop(['Table name','State code','Distt code'],axis=1)
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
dfS=df[df['Area name']!='INDIA']
```

```python
dfS.loc[:, 'Area name'] = dfS['Area name'].str.slice(start=8)
```

```python
dfS
```

```python
dfS.to_csv("dataset.csv")
```

```python
states=dfS['Area name'].unique()
```

```python
states=[item.title() for item in states]
```

```python
states.sort()
states
```

```python
for state in states:
    file_name="raw_data/"+state+".xlsx"
#     print(file_name)
    df=pd.read_excel(file_name,skiprows=6)
    df.columns=["Table name","State code","Distt code","Area name","Total/Rural/Urban",
           "Education level","total person","total males","total females","person main worker",
           "males main worker","females main worker","person marginal worker","males marginal worker",
            "females marginal worker","person MAW","males MAW","females MAW","person non worker",
            "males non worker","females non worker","person unemployed","males unemployed","females unemployed"]
    df=df.drop(['Table name','State code','Distt code'],axis=1)
    df=df.dropna()
    del_name='State - '+state.upper()
    df = df[df['Area name'] != del_name]
    df.loc[:, 'Area name'] = df['Area name'].str.slice(start=11)
    df['Area name']=df['Area name'].str.upper()
    df=df.dropna()
    df=df.reset_index(drop=True)
    df.to_csv('app/data/states_data/'+state+".csv")
```
